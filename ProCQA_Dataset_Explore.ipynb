Downloading ProCQA Dataset in Drive Directly


https://drive.google.com/drive/folders/1-7tTChmUuYCUzG58Vi462dhhKi-vcauR

```python
from google.colab import auth
from googleapiclient.discovery import build
from googleapiclient.http import MediaIoBaseDownload
import io
import os

# Authenticate and create the Google Drive client
auth.authenticate_user()
drive_service = build('drive', 'v3')

# Replace with the shared folder ID
shared_folder_id = "1-7tTChmUuYCUzG58Vi462dhhKi-vcauR"  # Change this to your folder ID

# Get the list of files in the shared folder
results = drive_service.files().list(
    q=f"'{shared_folder_id}' in parents and trashed=false",
    fields="files(id, name, mimeType)").execute()

files = results.get('files', [])

if not files:
    print("No files found in the shared folder.")
else:
    print(f"Found {len(files)} JSON files. Downloading...")

    # Mount Google Drive
    from google.colab import drive
    drive.mount('/content/drive')

    # Destination folder in Google Drive
    destination_folder = "/content/drive/My Drive/ProCQA_Dataset"
    os.makedirs(destination_folder, exist_ok=True)

    # Download each JSON file
    for file in files:
        file_id = file['id']
        file_name = file['name']
        mime_type = file['mimeType']

        if mime_type == "application/json" or file_name.endswith(".json"):
            request = drive_service.files().get_media(fileId=file_id)
            file_path = os.path.join(destination_folder, file_name)

            with open(file_path, "wb") as f:
                downloader = MediaIoBaseDownload(f, request)
                done = False
                while not done:
                    status, done = downloader.next_chunk()
                    print(f"Downloading {file_name}... {int(status.progress() * 100)}% done.")

    print(f"âœ… All JSON files downloaded to {destination_folder}")
```
